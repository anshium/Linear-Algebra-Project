# Recommender Systems

we will issustrate how Recommender Systems use different aspects of LA using visualisations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample ratings data
user_ids = [1, 1, 2, 2, 2, 3, 3, 4, 4, 4]
item_ids = [1, 2, 1, 3, 4, 2, 3, 1, 3, 4]
ratings = [4, 5, 3, 2, 4, 5, 1, 3, 4, 2]

ratings_data = pd.DataFrame({'user_id': user_ids, 'item_id': item_ids, 'rating': ratings})

# Rating Distribution
plt.hist(ratings_data['rating'], bins=10)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Rating Distribution')
plt.show()
```

This histogram visualizes the distribution of ratings given by users.
x-axis: Rating - It represents the different rating values.
y-axis: Frequency - It represents the number of occurrences for each rating value.
This plot helps us understand the distribution of ratings and provides an overview of the preferences of users in the dataset.

```python
# User-Item Matrix
user_item_matrix = ratings_data.pivot(index='user_id', columns='item_id', values='rating')
plt.imshow(user_item_matrix, cmap='hot', aspect='auto')
plt.xlabel('Item ID')
plt.ylabel('User ID')
plt.title('User-Item Matrix')
plt.colorbar()
plt.show()
```

This plot visualizes the user-item matrix, representing the interactions between users and items.
x-axis: Item ID - It represents the unique identifiers for each item in the dataset.
y-axis: User ID - It represents the unique identifiers for each user in the dataset.
Each cell represents the rating given by a user for an item, with different colors representing different rating values.This plot allows us to observe the pattern of interactions between users and items. It helps identify which items have been rated by which users and provides insights into the sparsity and distribution of the ratings.


```python
# Item Popularity
item_popularity = ratings_data['item_id'].value_counts()
plt.bar(item_popularity.index, item_popularity.values)
plt.xlabel('Item ID')
plt.ylabel('Number of Ratings')
plt.title('Item Popularity')
plt.show()
```

This bar plot illustrates the popularity of items based on the number of ratings they have received.
x-axis: Item ID - It represents the unique identifiers for each item in the dataset.
y-axis: Number of Ratings - It represents the count of ratings received by each item.
This plot helps us identify the popularity of items based on the number of ratings they have received. It provides insights into the distribution of ratings among different items and highlights items that are more frequently rated.

```python
# User Similarity Matrix
user_similarity_matrix = np.random.rand(len(user_item_matrix), len(user_item_matrix))
plt.imshow(user_similarity_matrix, cmap='coolwarm', aspect='auto')
plt.xlabel('User ID')
plt.ylabel('User ID')
plt.title('User Similarity Matrix')
plt.colorbar()
plt.show()
```

This heatmap represents the similarity between users based on their rating patterns.
x-axis: User ID - It represents the unique identifiers for each user in the dataset.
y-axis: User ID - It represents the unique identifiers for each user in the dataset.
Each cell represents the similarity score between two users, with colors indicating the degree of similarity.Each cell represents the similarity score between two users, with colors indicating the degree of similarity.

# Conclusion:

Based on the usage of linear algebra (LA) in the recommender system (RS), we can draw the following conclusions:

-User-Item Matrix: The user-item matrix, often represented as a matrix of ratings, is a fundamental component of RS. LA concepts, such as matrix operations and factorization techniques (e.g., singular value decomposition), are commonly used to analyze and manipulate the user-item matrix. These LA techniques enable efficient computations and help uncover latent factors or features that capture user-item interactions.

-Matrix Factorization: Matrix factorization methods, such as singular value decomposition (SVD) or non-negative matrix factorization (NMF), are widely used in RS. These techniques leverage LA to decompose the user-item matrix into lower-rank matrices or latent factor representations. By decomposing the matrix, RS can discover hidden patterns or features in the data, enabling accurate recommendations and dimensionality reduction.

-Similarity Metrics: LA provides various similarity metrics that are essential for RS. Cosine similarity, Euclidean distance, or Pearson correlation coefficient are commonly used LA-based measures to compute similarity between users or items. These metrics assist in identifying similar users or items based on their rating patterns, allowing for personalized recommendations.

-Collaborative Filtering: Collaborative filtering, a popular RS technique, relies on LA concepts to find similarities between users or items. LA operations, such as matrix multiplication or dot products, are used to calculate similarity scores and predict missing ratings. Collaborative filtering techniques leverage LA to recommend items to users based on the preferences of similar users or the relationships between items.

-Dimensionality Reduction: Dimensionality reduction techniques, like principal component analysis (PCA) or matrix factorization, are employed in RS to reduce the dimensionality of the user-item matrix. LA-based methods help identify the most informative dimensions or latent factors, enabling more efficient computations and improving recommendation accuracy.

In conclusion, the usage of linear algebra in recommender systems is vital for analyzing and manipulating the user-item matrix, performing matrix factorization, computing similarity metrics, implementing collaborative filtering, and applying dimensionality reduction techniques. These LA concepts play a crucial role in enhancing the efficiency, accuracy, and scalability of recommendation algorithms, enabling personalized and effective recommendations for users.
