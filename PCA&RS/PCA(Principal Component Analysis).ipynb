# Principal Component Analysis(PCA)

we will see an example how Linear Algebra is used in PCA for Machine Learning.
we will use iris dataset for this illustration.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Scatter plot of original data
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Original Data')
plt.show()
```

This plot visualizes the original Iris dataset in two dimensions.
x-axis: Sepal Length (cm) - It represents the sepal length measurements of the flowers.
y-axis: Sepal Width (cm) - It represents the sepal width measurements of the flowers.
Each species is represented by a different color.This plot provides an overview of the original data and allows us to observe the distribution and separation of the different Iris species based on sepal length and width.


```python
# Perform PCA with 2 components
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(X)

# Scatter plot of transformed data
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=y)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Transformed Data')
plt.show()
```

This plot displays the transformed Iris dataset after applying PCA.
x-axis: Principal Component 1 - It represents the first principal component obtained from PCA.
y-axis: Principal Component 2 - It represents the second principal component obtained from PCA.
The reduced-dimensional representation of the data using the principal components. It allows us to observe how the different species cluster or separate from each other based on the transformed features.


```python
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance Ratio')
plt.show()
```

This plot illustrates the explained variance ratio of each principal component.
x-axis: Number of Components - It represents the number of principal components included.
y-axis: Cumulative Explained Variance - It represents the cumulative explained variance ratio.
This plot shows the cumulative amount of variance explained by each additional principal component. It helps determine the number of principal components to retain based on the amount of information they capture.

```python
# Scree plot
eigenvalues = pca.explained_variance_
plt.plot(range(1, len(eigenvalues)+1), eigenvalues, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Eigenvalues')
plt.title('Scree Plot')
plt.show()
```

This plot displays the eigenvalues of each principal component.
x-axis: Number of Components - It represents the number of principal components included.
y-axis: Eigenvalues - It represents the magnitude of each eigenvalue.
This plot helps determine the number of significant principal components by observing the eigenvalues. The eigenvalues represent the amount of variance explained by each principal component.

```python
feature_names = iris.feature_names
pcs = pca.components_.T
plt.figure(figsize=(8, 8))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=y)
for i, feature in enumerate(feature_names):
    plt.arrow(0, 0, pcs[i, 0], pcs[i, 1], color='r', alpha=0.75)
    plt.text(pcs[i, 0] * 1.2, pcs[i, 1] * 1.2, feature, color='b', ha='center', va='center')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Biplot')
plt.show()
```

This plot combines the scatter plot of the transformed data with arrows representing the loadings of the original features.Data points are Same as the scatter plot of transformed data.Each arrow represents the loading (contribution) of an original feature to the principal components.The biplot allows us to visualize both the data points and the relationship between the principal components and original features. It helps understand which features have the most influence on the principal components and how the data points are aligned with these features.

# Conclusion:

Based on the visualizations obtained from the PCA analysis of the Iris dataset, we can draw the following conclusions:

-Dimensionality Reduction: PCA effectively reduces the dimensionality of the dataset by transforming the original features into a new set of uncorrelated variables called principal components. By applying PCA, we can visualize the data in a lower-dimensional space while preserving much of the important information.

-Data Separability: The scatter plot of the transformed data demonstrates that the different species of Iris flowers exhibit distinct patterns and clusters in the reduced-dimensional space. This suggests that the principal components derived from PCA are able to capture meaningful variations in the data and help separate different classes.

-Explained Variance: The explained variance ratio plot shows the proportion of the total variance explained by each principal component. In this case, the first two principal components explain a significant amount of the variance in the Iris dataset. The cumulative explained variance plot indicates that the first two principal components capture approximately 95% of the total variance. Therefore, by retaining these two components, we can represent a large portion of the original data.

-Feature Loadings: The biplot allows us to understand the relationship between the original features and the principal components. The arrows in the biplot represent the loadings or contributions of each feature to the principal components. Longer arrows indicate stronger contributions. By examining the biplot, we can observe which features have the most influence on the principal components and their relative orientations.

Overall, PCA provides a valuable tool for dimensionality reduction, visualization, and understanding the underlying structure of the data. It helps identify the most important features, visualize data separability, and provide insights into the relationships between variables. In the case of the Iris dataset, PCA successfully captures the main patterns and clusters within the data, making it easier to analyze and interpret.
